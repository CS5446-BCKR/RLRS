```python
def unnest_modoptions(row):
    option_ids = str(row['mod_id']).split('|')
    return pd.DataFrame({'order_id': [row['order_id']] * len(option_ids),
                         'item_id': [row['item_id']] * len(option_ids),
                         'customer_id': [row['customer_id']] * len(option_ids),
                         'created_at': [row['created_at']] * len(option_ids),
                         'served_at	': [row['served_at']] * len(option_ids),
                         'option_id': option_ids})
```

```python
order_item_mods = pd.concat([unnest_modoptions(row) for _, row in order_item.iterrows()], ignore_index=True)
```

```python
# below is the example of how to join back to mod options
order_item_mods.merge(mod_options,left_on = 'option_id', right_on = 'id',how='left')
```

```python
order_item_mods.rename(columns={'served_at\t':'serve_at'},inplace=True)
```

```python
order_item_mods.to_csv(f'{RLRS_root_location}/data/ayampp/order_item_mods.csv',header=True,index=False)
```

```python
order_item_mods.columns
```

```python
order_item_mods.created_at = pd.to_datetime(order_item_mods.created_at, unit='ms')
```

```python
# Group by 'customer_id', 'item_id', and 'option_id' and aggregate to get order frequency
grouped_df = order_item_mods.groupby(['customer_id', 'item_id', 'option_id']).size().reset_index(name='frequency')
```

```python
# Select the last 'created_at' timestamp for each user's order
last_order_timestamps = order_item_mods.groupby('customer_id')['created_at'].max().reset_index()
```

```python
# Merge order frequency and last order timestamp dataframes
merged_df = pd.merge(grouped_df, last_order_timestamps, on='customer_id')
```

```python
merged_df['item_option_id'] = merged_df.item_id + '-' + merged_df.option_id
```

```python
# final output
merged_df[["customer_id","item_option_id","frequency","created_at"]].to_csv(f'{RLRS_root_location}/data/ayampp/user_item_frequency_table.csv',header=True,index=False)
```
